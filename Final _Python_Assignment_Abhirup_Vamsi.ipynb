```python
## this is the function which shall read the data and make histograms out of the numeric data sets
## the output of the data is loaded into .png format using the plt.savefig function
## the split of the numeric and categorical values is done using the .select_dtypes methods

def Histogram(data):
    #data.select_dtypes(include=['int64'])
    
    data_catgorical = data.select_dtypes(include=['object']) #passing the categorical columns into data_categorical
    data_numerical = data.select_dtypes(exclude=['object']) # passing the numerical columns into data_numerical
    
    for i in list(data_numerical.columns): 
        fig = plt.figure(figsize=(10,10))
        
        sns.distplot( data_numerical[i] , color="red", label=str(i),bins=20, kde=False) #plotted the distplot
        plt.savefig(i+".png", orientation='Landscape',transparent=True,edgecolor='g') #saved the fig
```

```python
## the function can take the directory or the repository in which the user wants to save the .png files
## the argument variable declared here is dir.

def Graphs(data, var, dir):
    #data.select_dtypes(include=['int64'])
    
    os.chdir(dir)
    data2 = data.iloc[:, var]
    data_catgorical = data2.select_dtypes(include=['object']) 
    data_numerical = data2.select_dtypes(exclude=['object'])
    
    
    for i in list(data_numerical.columns):
        fig, ax = plt.subplots(1,2, figsize = (10,10))
        sns.distplot( data_numerical[i] , color="red", label=str(i),bins=20, kde=False, ax= ax[0])
        sns.boxplot( data_numerical[i],ax= ax[1])
        plt.savefig(i+"4.png", ax = ax[1])
        #plt.show()
        
        
    for j in list(data_catgorical.columns):     
        plt.figure(figsize = (10,10))
        sns.countplot(x = data_catgorical[j], data = data_catgorical )
        plt.savefig(j+"4.png")
        #plt.show()
```

```python
## we are doing two enhancements. 
## First, in case there are two or more continuous variables analysed, the last if statement will generate a 
## correlation heatmap which is a very helpful tool in doing feature engineering and bivariate analysis.


## second, there are some cases where the variable is shown as a variable is actually is a continuos data. 
## we have added an attribute check. In case the user want to convert some of the variables from numeric to 
## object type variables, he/she can do so passing values in the check list.


def Graphs(data, var, dir, check):
    #data.select_dtypes(include=['int64'])
    
    os.chdir(dir)
    
    for k in check:
        data[k] = data[k].astype('object')
    
    data2 = data.iloc[:, var]
    data_catgorical = data2.select_dtypes(include=['object']) 
    data_numerical = data2.select_dtypes(exclude=['object'])
    
    
    for i in list(data_numerical.columns):
        fig, ax = plt.subplots(1,2, figsize = (10,10))
        sns.distplot( data_numerical[i] , color="red", label=str(i),bins=20, kde=False, ax= ax[0])
        sns.boxplot( data_numerical[i],ax= ax[1])
        plt.savefig(i+"5.png", ax = ax[1])
        #plt.show()
        
        
    for j in list(data_catgorical.columns):     
        plt.figure(figsize = (10,10))
        sns.countplot(x = data_catgorical[j], data = data_catgorical )
        plt.savefig(j+"5.png")
        #plt.show()
        
        
    ##Feature 2:  Create a set of code which shall return a correlation matrix in case there are more than 2 numerical variables
    
    if data_numerical.shape[1] >= 2:
        plt.figure(figsize = (10,10))
        sns.heatmap(data_numerical.corr(),annot=True,
        square=True, linewidths=.5, cbar_kws={"shrink": .5})
        plt.savefig("heatmap"+"5.png")
        #plt.show()
```
